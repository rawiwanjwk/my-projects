```python
# TODO 04 - your friend ask for all order data in `California` and `Texas` in 2017 (look at Order Date), send him csv file

# df[ (df['State'] == "California") | (df['State'] == "Texas")] this way can be used
result2 = df.query('State == ["California","Texas"]')

data_2017 = result2[result2['Order Date'].dt.year == 2017]

data_2017.to_csv("data_2017")
data_2017
```

```python
# TODO 05 - how much total sales, average sales, and standard deviation of sales your company make in 2017
result3 = df[df['Order Date'].dt.year == 2017]
agg_result3 = result3['Sales'].agg(['sum', 'mean', 'std'])
agg_result3
```

```python
# TODO 06 - which Segment has the highest profit in 2018
df[df['Order Date'].dt.year == 2018]\
    .groupby('Segment')['Profit'].sum()\
        .sort_values(ascending=False)
```

```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019
df[df['Order Date'].between('2019-04-15', '2019-12-31')]\
    .groupby('State')['Sales'].sum()\
        .sort_values(ascending=True)\
            .head(5)
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25%
sales_2019 = df[df['Order Date'].dt.year == 2019].groupby('Region')['Sales'].sum()
sum_sales = sales_2019.sum()
portion_wc = (sales_2019['West'] + sales_2019['Central'])/sum_sales
portion_wc
```

```python
# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020
df[df['Order Date'].between('2019-01-01', '2020-12-31')]\
    .groupby('Product Name')[['Quantity', 'Sales']].sum()\
        .sort_values(['Quantity', 'Sales'], ascending=False).head(10)
```
