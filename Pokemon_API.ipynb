# **API**

```python
import requests
import pandas as pd

pokemon = []

for i in range(5):
    url = f" https://pokeapi.co/api/v2/pokemon/{i+5}"
    response = requests.get(url)
    data = response.json()
    poke_name = data["name"]
    poke_weight = data["weight"]
    poke_height = data["height"]
    poke_type = data["types"][0]["type"]["name"]
    poke_ability = data["abilities"][0]["ability"]["name"], data["abilities"][1]["ability"]["name"]
    pokemon.append({
        "Name": poke_name,
        "Weight": poke_weight,
        "Height": poke_height,
        "Type": poke_type,
        "Abilities": poke_ability
    })

print(pd.DataFrame(pokemon))



```
