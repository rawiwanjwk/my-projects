# **Hammer Scissors Paper Game**


```python
import random
def hands_game():
    print("Hello Player!")
    player_score = 0
    bot_score = 0
    round = 0
    hands = ["hammer", "scissors", "paper"]
    player_name = input("What's your name?: ").upper()
    print(f"Welcome to Hands game {player_name}!")
    print("Rule: write the hand you choose 'hammer','scissors','paper' or 'stop'")
    while True:
        player_hand = input("Choose your hand or `stop`: ").lower()
        bot_hand = random.choice(hands)
        if player_hand == "stop":
            print(f"total round: {round}")
            print(f"{player_name} score: {player_score}")
            print(f"Bot score: {bot_score}")

            if player_score > bot_score:
                print("You Win!")
            elif player_score < bot_score:
                print("You Lose!")
            else:
                print("Even Score!")
            break
        elif player_hand not in hands or not "stop":
            print("Invalid Value, Please try again!")
            continue


        ## player win
        if (player_hand == "hammer" and bot_hand == "scissors") or (player_hand == "scissors" and bot_hand == "paper") or (player_hand == "paper" and bot_hand == "hammer"):
            print(f"Bot_hand: {bot_hand}")
            player_score += 1
            bot_score += 0
            round += 1
            print(f"Round: {round} Player WIN!")
            print(f"Player Score: {player_score}")
            print(f"Bot Score: {bot_score}")
        ## even
        elif player_hand == bot_hand:
            print(f"Bot_hand: {bot_hand}")
            player_score += 0
            bot_score += 0
            round += 1
            print(f"Round: {round} BOTH EVEN!")
            print(f"Player Score: {player_score}")
            print(f"Bot Score: {bot_score}")
        ## player lose
        elif (player_hand == "scissors" and bot_hand == "hammer") or (player_hand == "paper" and bot_hand == "scissors") or (player_hand == "hammer" and bot_hand == "paper"):
            print(f"Bot_hand: {bot_hand}")
            player_score += 0
            bot_score += 1
            round += 1
            print(f"Round: {round} Player LOSE!")
            print(f"Player Score: {player_score}")
            print(f"Bot Score: {bot_score}")
```

## **Result**

```python
hands_game()
```

# **Pizza Restaurant**

```python
# create pizza menu dict
import pandas as pd
import numpy as np

pizza_menu = [
    {"id": 1,
    "category": "Pizza",
    "name": "Hawaiian pizza",
    "size&price":{"M": 200,
                  "L": 300,
                  "XL": 450}},
    {"id": 2,
    "category": "Pizza",
    "name": "Double cheese pizza",
    "size&price": {"M": 200,
                  "L": 300,
                  "XL": 400}},
    {"id": 3,
    "category": "Pizza",
    "name": "Truffle pizza",
    "size&price": {"M": 250,
                  "L": 370,
                  "XL": 480}},
    {"id": 4,
    "category": "Drink",
    "name": "Coke",
    "size&price": {"S": 20,
                  "M" : 35}},
    {"id": 5,
    "category": "Drink",
    "name": "Pepsi",
    "size&price": {"S": 20,
                  "M": 35}},
    {"id": 6,
    "category": "Drink",
    "name": "Still water",
    "size&price": {"S": 15,
                    "M": 30}}
]

def pizza_hat():
    total_bill = 0
    order_table = {}
    taking_order = True
    print("Welcome to Pizza Hat Restaurant")
    print("Promotion: Get 10% discount if you spend >=1,000 THB üòò")
    print("Here's our menu")
    print(pd.DataFrame(pizza_menu))
    while (taking_order == True):
        order = input("What would you like to eat? write 'id number(1-6)' or 'done': ").lower()
        if (order == 'done'):
            if not order_table:
                print("You didn't order anything. Thank you for visiting")
            else:
                print("\n" + "="*30)
                print("       YOUR RECEIPT üßæ     ")
                print("="*30)
                # ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏Ñ‡∏¥‡∏î‡πÄ‡∏á‡∏¥‡∏ô
                for name, sizes in order_table.items():
                    for item in pizza_menu: # ‡πÄ‡∏£‡∏≤‡∏ï‡πâ‡∏≠‡∏á‡πÑ‡∏õ‡∏´‡∏≤ 'item' ‡πÉ‡∏ô pizza_menu ‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏ä‡∏∑‡πà‡∏≠‡∏ï‡∏£‡∏á‡∏Å‡∏±‡∏ö name ‡∏Å‡πà‡∏≠‡∏ô
                        if item["name"] == name:
                            for size, item_amount in sizes.items():
                                price = item["size&price"][size]
                                subtotal = price * item_amount
                                total_bill += subtotal
                                print(f" {name} ({size}) {price} x {item_amount}: {price * item_amount:,.2f} THB")
                print("-" * 30)
                print(f"Subtotal: {total_bill:,.2f} THB")
                # --- ‡∏™‡πà‡∏ß‡∏ô‡∏•‡∏î 10% ‡πÄ‡∏°‡∏∑‡πà‡∏≠‡∏ã‡∏∑‡πâ‡∏≠‡∏Ñ‡∏£‡∏ö 1,000 ‡∏ö‡∏≤‡∏ó ---
                if total_bill >= 1000:
                    discount = total_bill * 0.10
                    final_price = total_bill - discount
                    print(f"Discount (10%): -{discount} THB")
                else:
                    final_price = total_bill

                print(f"Total amount to pay: {final_price:,.2f} THB")
                print("="*30)
                print("Thank you for your order! üçï")
            taking_order == False
            break
        if not order.isdigit(): # ‡πÄ‡∏ä‡πá‡∏Ñ‡∏ß‡πà‡∏≤‡∏•‡∏π‡∏Å‡∏Ñ‡πâ‡∏≤‡πÉ‡∏™‡πà‡πÄ‡∏•‡∏Ç id ‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏ñ‡∏π‡∏Å‡πÑ‡∏´‡∏°
            print("‚ùå Please enter a valid menu ID number(1-6) or 'done'.")
            continue

        # check_valid_order
        elif (order != 'done'):
            order_id = int(order) # ‡πÅ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç
            found = False # ‡∏ï‡∏±‡πâ‡∏á‡∏™‡∏ñ‡∏≤‡∏ô‡∏∞‡∏ß‡πà‡∏≤‡∏¢‡∏±‡∏á‡πÑ‡∏°‡πà‡πÄ‡∏à‡∏≠‡πÄ‡∏°‡∏ô‡∏π
            for item in pizza_menu:
                if item["id"] == order_id:
                    found = True
                    name = item["name"]    # ‡∏î‡∏∂‡∏á‡∏ä‡∏∑‡πà‡∏≠‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏°‡∏≤‡πÄ‡∏õ‡πá‡∏ô Key ‡πÉ‡∏ô dict
                    print(f"You choose: {name}")
                    print(f"Available sizes for {name}: {list(item['size&price'].keys())}")
                    while True:  # ‡πÄ‡∏ä‡πá‡∏Ñ‡∏ß‡πà‡∏≤‡∏•‡∏π‡∏Å‡∏Ñ‡πâ‡∏≤‡πÉ‡∏™‡πà‡πÑ‡∏ã‡∏™‡πå‡∏ñ‡∏π‡∏Å‡πÑ‡∏´‡∏°
                        size = input("What size would you like?: ").upper()
                        if size not in item["size&price"]:
                            print("Sorry, this size is not available for this item. please try again!")
                        else:
                            break
                    while True: # ‡πÄ‡∏ä‡πá‡∏Ñ‡∏ß‡πà‡∏≤‡∏•‡∏π‡∏Å‡∏Ñ‡πâ‡∏≤‡πÉ‡∏™‡πà‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏ñ‡∏π‡∏Å‡πÑ‡∏´‡∏°
                        item_amount = int(input("How many items do you want?: "))
                        if item_amount <= 0:
                            print("Sorry, Please put positive number.")
                        else:
                            break
                    if name not in order_table: # 1. ‡πÄ‡∏ä‡πá‡∏Ñ‡∏ß‡πà‡∏≤‡∏°‡∏µ‡∏ä‡∏∑‡πà‡∏≠‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡πÉ‡∏ô‡πÇ‡∏ï‡πä‡∏∞‡∏™‡∏±‡πà‡∏á‡∏≠‡∏≤‡∏´‡∏≤‡∏£‡∏´‡∏¢‡∏±‡∏á
                        order_table[name] = {} # ‡∏ñ‡πâ‡∏≤‡∏¢‡∏±‡∏á‡πÑ‡∏°‡πà‡∏°‡∏µ ‡πÉ‡∏´‡πâ‡∏™‡∏£‡πâ‡∏≤‡∏á Dict ‡∏ß‡πà‡∏≤‡∏á‡∏£‡∏≠‡πÑ‡∏ß‡πâ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÄ‡∏Å‡πá‡∏ö‡πÑ‡∏ã‡∏™‡πå
                    if size in order_table[name]: # 2. ‡πÄ‡∏ä‡πá‡∏Ñ‡∏ß‡πà‡∏≤‡∏†‡∏≤‡∏¢‡πÉ‡∏ï‡πâ‡∏ä‡∏∑‡πà‡∏≠‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏ô‡∏±‡πâ‡∏ô ‡∏°‡∏µ‡∏Å‡∏≤‡∏£‡∏™‡∏±‡πà‡∏á‡πÑ‡∏ã‡∏™‡πå‡∏ô‡∏µ‡πâ‡πÑ‡∏õ‡∏¢‡∏±‡∏á
                        order_table[name][size] += item_amount # ‡∏ñ‡πâ‡∏≤‡πÄ‡∏Ñ‡∏¢‡∏™‡∏±‡πà‡∏á‡πÑ‡∏ã‡∏™‡πå‡∏ô‡∏µ‡πâ‡πÅ‡∏•‡πâ‡∏ß ‡πÉ‡∏´‡πâ‡∏ö‡∏ß‡∏Å‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡πÄ‡∏û‡∏¥‡πà‡∏°
                    else:
                        order_table[name][size] = item_amount # ‡∏ñ‡πâ‡∏≤‡∏¢‡∏±‡∏á‡πÑ‡∏°‡πà‡πÄ‡∏Ñ‡∏¢‡∏™‡∏±‡πà‡∏á‡πÑ‡∏ã‡∏™‡πå‡∏ô‡∏µ‡πâ ‡πÉ‡∏´‡πâ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏Ñ‡πà‡∏≤‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô
                    break
            if not found:
                print(f"‚ùå ID {order_id} not found in our menu. Please try again!")
```
